```python
fp = '/home/blackmamba/code/pbragamiranda/presidential-speeches-scraper/bolsonaro_data/csv_data/0a9e579de17e4f70b0ffa08469937286___09122019.csv'
df1 = pd.read_csv(fp, sep='|')
df1['speech']= df1['speech'].astype(str)
df1['title']= df1['title'].astype(str)
```

```python
lower_alpha = lambda x: x.lower()
df1['speech'] = df1.speech.map(lower_alpha)
```

```python
import string

punc_re = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
df1['speech'] = df1.speech.map(punc_re)

df1.head()
```

```python
# tokenization reviews
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

df1['tokens'] = df1.speech.map(word_tokenize)
df1.head()
```

```python
# creating a list combining all tokens
word_list = sum(df1.tokens.tolist(), [])
word_list[:10]
```

```python
# finding the most commun words

from collections import Counter

# converting the list in a dictionary with values count
word_counts = Counter(word_list)

# reverting the key / values in the dictionary to be able to sort
word_counts = list(zip(word_counts.values(), word_counts.keys()))

# sorting the list by count
word_counts = sorted(word_counts, reverse=True)

# print
word_counts[:10]
```

```python
# Removing the stop words
from nltk.corpus import stopwords
nltk.download("stopwords")

stop_words = stopwords.words('portuguese')

stop_lambda = lambda x: [y for y in x if y not in stop_words]
df1['tokens_stop'] = df1.tokens.apply(stop_lambda)
df1.head()
```
